```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector, array_to_latex, plot_state_qsphere
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(4, 2)
circuit.x([0,1])
circuit.barrier()
circuit.cx(0, 2)
circuit.cx(1, 2)
circuit.ccx(0,1, 3)
circuit.barrier()
```

```python
simulator = Aer.get_backend("statevector_simulator")
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
array_to_latex(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python
circuit.measure([2,3], [0,1])
circuit.draw(output="mpl")
```

```python
# simulate circuit and print message with SUM and CARRY-OUT states
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1).result()
counts = result.get_counts()
print(f'SUM is {list(counts.keys())[0][1]}')
print(f'CARRY-OUT is {list(counts.keys())[0][0]}')
```
