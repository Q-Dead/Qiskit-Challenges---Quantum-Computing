```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector, plot_state_qsphere, array_to_latex
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(2)
circuit.h(0)
circuit.cx(0,1)
circuit.measure_all()
circuit.draw(output="mpl")
```

```python
simulator = Aer.get_backend("qasm_simulator")
result = execute(circuit, backend = simulator).result()
plot_histogram(result.get_counts())
```

```python
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.tools.monitor import job_monitor
IBMQ.save_account() # Private API Key
provider = IBMQ.load_account()
```

```python
qcomp = provider.get_backend('ibm_lagos')
job = execute(circuit, backend = qcomp)
job_monitor(job)
```

```python
qcomp_result = job.result()
plot_histogram(qcomp_result.get_counts())
```
